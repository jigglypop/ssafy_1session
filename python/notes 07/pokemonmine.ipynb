```python
import random

class Pokemon:
    
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        print(f'{self.name}의 HP가 {self.hp}로 설정되었습니다.')
        
    def body_attack(self, enemy):
        print(f'{self.name}이 {enemy.name}을 몸통박치기 공격을 했다....!')
        if random.randint(0, 1):
            enemy.hp -= self.level * 5
            print(f'{enemy.name}의 체력이 {enemy.hp}가 되었다....!')
        else:
            print('공격에 실패했다.')
            
    def bark(self):
        return print(f'{self.name}: 우렁찬 울음소리 pikachu!')
        
    def thousond_volt(self,enemy):
        print(f'{self.name}이 {enemy.name}을 100만볼트로 공격했다...!')
        if random.randint(0,1):
            enemy.hp -= self.level * 7
            print(f'{enemy.name}의 체력이 {enemy.hp}가 되었다...!')
        else:
            print('공격에 실패했다.')
            
    def __repr__(self):
        return f'{self.name}{self.hp}: lv.{self.level} exp.{self.exp}/100'
```

```python
p1 = Pokemon('피카츄1')
p2 = Pokemon('피카츄2')
p1.body_attack(p2)
p1.bark()
p1.thousond_volt(p2)
```

```python
import random



turn = 0
p1_name = input('플레이어 1의 이름을 정해주세요: ')
p2_name = input('플레이어 2의 이름을 정해주세요: ')
p1 = Pokemon(p1_name)
p2 = Pokemon(p2_name)
print('===========게임이 시작되었습니다================')
while True:
    print(f'==========={turn}================')
    a = random.randint(0,2)
    if a == 0:
        p1.bark()
        p2.bark()
    elif a == 1:
        if turn % 2:
            p1.body_attack(p2)
            print(p1)
            print(p2)
        else:
            p2.body_attack(p1)
            print(p1)
            print(p2)
    elif a == 2:
        if turn % 2:
            p1.thousond_volt(p2)
            print(p1)
            print(p2)
        else:
            p2.thousond_volt(p1)
            print(p1)
            print(p2)        
    if p1.hp < 0 or p2.hp < 0:
        break
    turn += 1
```
